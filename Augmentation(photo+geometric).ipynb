```python
#augmentation for Botrytis Blight 

import os
from os import listdir
from os.path import isfile, join
import numpy as np
import cv2 
from scipy import fftpack
from PIL import Image,ImageEnhance
import scipy.fftpack
from skimage import data, io
from scipy.signal import convolve2d as conv2



folder1 = '/content/drive/MyDrive/Onion Diseases/Raw image/Botrytis Blight'

dirr1 = '/content/drive/MyDrive/Onion Diseases/Augmentation/Botrytis Blight'


j=1


for im in os.listdir(folder1):

    #read image
    path = os.path.join(folder1,im)
    img = cv2.imread(path)
    cv2.imwrite(os.path.join(dirr1,'raw '+str(j)+'.png'),img)
    imgg = Image.open(path)
   
    #augmentation
    f1= cv2.flip(img,0)
    cv2.imwrite(os.path.join(dirr1,'flip1 '+str(j)+'.png'),f1)
    f2= cv2.flip(img,-1)
    cv2.imwrite(os.path.join(dirr1,'flip2 '+str(j)+'.png'),f2)
    f3= cv2.flip(img,1)
    cv2.imwrite(os.path.join(dirr1,'flip3 '+str(j)+'.png'),f3)
    f4 = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)
    cv2.imwrite(os.path.join(dirr1,'rotate1 '+str(j)+'.png'),f4)
    f5 = img[30:194,30:194]
    cv2.imwrite(os.path.join(dirr1,'zoom1 '+str(j)+'.png'),f5)
    f6 = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
    cv2.imwrite(os.path.join(dirr1,'rotate2 '+str(j)+'.png'),f6)
    f7 = cv2.rotate(img, cv2.ROTATE_180)
    cv2.imwrite(os.path.join(dirr1,'rotate3 '+str(j)+'.png'),f7)
    f8 = img[40:164,40:164]
    cv2.imwrite(os.path.join(dirr1,'zoom2 '+str(j)+'.png'),f8)
    
    #brightness
    enhancer = ImageEnhance.Brightness(imgg)
    factor = 0.5 #darkens the image
    im_output = enhancer.enhance(factor)
    im_output.save(os.path.join(dirr1,'br_low '+str(j)+'.png'))
    factor = 1.5 #brightens the image
    im_output = enhancer.enhance(factor)
    im_output.save(os.path.join(dirr1,'br_high '+str(j)+'.png'))

    #color
    enhancer = ImageEnhance.Color(imgg)
    factor = 0.5 #darkens the image
    im_output = enhancer.enhance(factor)
    im_output.save(os.path.join(dirr1,'color1 '+str(j)+'.png'))
    factor = 1.5 #brightens the image
    im_output = enhancer.enhance(factor)
    im_output.save(os.path.join(dirr1,'color2 '+str(j)+'.png'))

    #Sharpness 
    enhancer = ImageEnhance.Sharpness(imgg)
    factor = 0.8 #darkens the image
    im_output = enhancer.enhance(factor)
    im_output.save(os.path.join(dirr1,'Sharpness1 '+str(j)+'.png'))
    factor = 1.8 #brightens the image
    im_output = enhancer.enhance(factor)
    im_output.save(os.path.join(dirr1,'Sharpness2 '+str(j)+'.png'))

    #contrast
    enhancer2 = ImageEnhance.Contrast(imgg)
    factor = 0.5 #darkens the image
    im_output = enhancer2.enhance(factor)
    im_output.save(os.path.join(dirr1,'cntr_low '+str(j)+'.png'))
    factor = 1.5 #brightens the image
    im_output = enhancer2.enhance(factor)
    im_output.save(os.path.join(dirr1,'cntr_high '+str(j)+'.png'))
   
    j=j+1
```
